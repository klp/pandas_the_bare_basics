```python
# create subset 
woodside_bools = df.neighbourhood == 'Woodside' # boolean list
woodside_bools
```

```python
# Create little data frames jammed together, then run average and collapses group by
df.groupby(['neighbourhood_group']) # creates five little dataframes by neighbourhood group
```

```python
# Need to collapse it again to get something useful
df.groupby(['neighbourhood_group']).price.mean()
```

```python
# Sort in order
df.groupby(['neighbourhood_group']).price.mean().sort_values()
```

```python
# Save series to variable
gentrification = df.groupby(['neighbourhood_group']).price.mean().sort_values()
```

```python
# make a chart
gentrification.plot(kind='bar')
```

```python
group_and_hood = df.groupby(['neighbourhood_group','neighbourhood']).price.mean().sort_values()
group_and_hood
```

```python
hood = df.groupby(['neighbourhood']).price.mean().sort_values()

hood = hood.head(10) # 10 cheapest neighborhoods

hood
```

```python
hood.plot(kind='bar')
```

```python
hood = df.groupby(['neighbourhood']).price.mean().sort_values()

hood = hood.tail(10) # 10 cheapest neighborhoods

hood
```

```python
hood.plot(kind='bar')
```
